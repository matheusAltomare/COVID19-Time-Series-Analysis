```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import os
```

```python
df_orig = pd.read_csv("https://brasil.io/dataset/covid19/caso?format=csv")
```

```python
df = df_orig.copy()
df
```

```python
#key = 'confirmed'
df_dict = {}
keys = ['confirmed', 'deaths']
for key in keys:
    df_new = pd.DataFrame()
    for state in df['state'].unique():
        data = df.groupby('state').get_group(state).groupby('date').sum()[[key]]
        data.columns = [state]
        df_new = pd.concat([df_new, data], axis=1)
    df_new.sort_index(inplace=True)
    df_dict[key] = df_new
```

```python
population_per_million = df.groupby('state').sum()['estimated_population_2019']/1e06
```

```python
for key in keys:
    last_day = (df_dict[key]/population_per_million).iloc[-1:].T
    last_day.sort_values(by=last_day.columns[0], ascending=False, inplace=True)
    last_day.plot(figsize=(14,5), 
                  kind='bar', 
                  title=f'{key.title()}/Population per million', 
                  grid=True,
                  fontsize=12,
                  rot=0)
    try:
        plt.savefig(f'results_brazil/all_states_{key}_normalized_population.png', transparent=True)
    except:
        os.system('mkdir results_brazil')
        plt.savefig(f'results_brazil/all_states_{key}_normalized_population.png', transparent=True)
        
    plt.show()
```

```python
states = ['RJ', 'SP', 'AM']

for key in keys:
    (df_dict[key][states]/population_per_million[states]).plot(figsize=(7,5), 
                                                              logy=True, 
                                                              title=f'{key.title()}/Population per million', 
                                                              grid=True,
                                                              fontsize=12)
    
    try:
        plt.savefig(f'results_brazil/{key}_normalized_population.png', transparent=True)
    except:
        os.system('mkdir results_brazil')
        plt.savefig(f'results_brazil/{key}_normalized_population.png', transparent=True)
        
    plt.show()
```
